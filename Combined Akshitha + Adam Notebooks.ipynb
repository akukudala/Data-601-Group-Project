## Adam's code:  Used to analyze growth vs spending

```python
file = "Cleaned Graduation Factors Data for MD, VA, DC by year, 2015-2021.xlsx"

# During the process, it was revealed that 2017 & 2018 data are identical for graduation rates.  They are 
# consequently left out of the analysis.
```

```python
import pandas as pd
import numpy

####Start with a bunch of cleanup stuff:

#Specific to the problem at hand
sheets = [1,2,3,4]
columns = ["NAME", "County", "December", "Locale*", "County Value", "PPCSTOT"]

df = pd.read_excel(file, sheet_name = sheets, header = 1, usecols = columns)

# Renames columns to be specific to the year they're in
for sheet in sheets:
    add = str(int(df[sheet]["December"][0]))
    df[sheet].rename(columns = {
        "December":"Unemployment Rate " + add, 
        "County Value": "Grad Rate "+add, 
        "PPCSTOT": "PPCSTOT " + add,
    }, inplace = True)
    df[sheet].drop(0, inplace=True)

    
# Creates one dataframe with all of hte data we need
joindf = pd.merge(df[1],df[2])
joindf = pd.merge(joindf,df[3])
joindf = pd.merge(joindf,df[4])

# Drops random empty lines that exist for some reason:  There were 20 lines? Idk why.
# Keeps all lines that have at least one "not na", fixes all indexes to contain no gaps
joindf = joindf[joindf.notna().any(axis=1)]
joindf.reset_index(drop = True, inplace = True)



joindf.sample(5)
```

```python
print(joindf.iloc[1])
```

```python
# All data that includes an NaN at all
joindf[joindf.isna().any(axis=1)]
```

```python
# Looks like we can drop the N/As bc they're almost all in grad rate, which is what we need for this analysis

# Drops any N/A rows and fixed indexing
joindf.dropna(inplace=True)
joindf.reset_index(drop = True, inplace = True)

joindf.sample(5)
```

```python
# Create the new data about growth rates if we have anything helpful
# Also this is some lovely terrible coding

growthdf = {'Year':[], 'County':[], 'Growth':[], 'Spending':[], 'Unemployment':[], 'Graduation Rate':[]}


numRows = joindf.shape[0]
for row in range(numRows):
    try:
        # Save as 2015 (when the money was spent), compare 2016/2015
        growth = 100 * (joindf['Grad Rate 2016'][row] - joindf['Grad Rate 2015'][row])

        if not numpy.isnan(growth):
            growthdf['Year'].append('2015')
            growthdf['County'].append(joindf['NAME'][row])
            growthdf['Growth'].append(growth)
            growthdf['Spending'].append(joindf['PPCSTOT 2015'][row])
            growthdf['Unemployment'].append(joindf['Unemployment Rate 2015'][row])
            growthdf['Graduation Rate'].append(joindf['Grad Rate 2015'][row])
        else:
            print(joindf.iloc[row])
            
        # Save as 2016 (when the money was spent), compare 2017/2016
        growth = 100 * (joindf['Grad Rate 2017'][row] - joindf['Grad Rate 2016'][row])

        if not numpy.isnan(growth):
            growthdf['Year'].append('2016')
            growthdf['County'].append(joindf['NAME'][row])
            growthdf['Growth'].append(growth)
            growthdf['Spending'].append(joindf['PPCSTOT 2016'][row])
            growthdf['Unemployment'].append(joindf['Unemployment Rate 2016'][row])
            growthdf['Graduation Rate'].append(joindf['Grad Rate 2016'][row])
        else:
            print(joindf.iloc[row])
             
    except:
        print("missing data", ", ", row,"\n", joindf.iloc[row])
        
growthdf = pd.DataFrame(growthdf)
growthdf.sample(5)
```

```python
# Visualization of the data taken from Akshitha's coding setup
import plotly
import plotly.express as px
```

```python
fig = px.scatter(
    growthdf,
    title="Comparison between Spending and Graduation Rate/Growth",
    x="Spending",
    y="Graduation Rate",
    # size="County Value",
    color="Growth",
    template="plotly_dark",
)
# fig.update_traces(textposition='top left')
# Years 
fig.update_layout(showlegend= False)
fig.show()
```

```python
fig = px.scatter(
    growthdf,
    title="Comparison between Spending and Graduation Rate Growth",
    x="Spending",
    y="Growth",
    # size="County Value",
    color="Graduation Rate",
    template="plotly_dark",
)
# fig.update_traces(textposition='top left')
# Years 
fig.update_layout(showlegend= False)
fig.show()
```

```python
# Shows average yearly growth.  This line showed that 2017 experienced 0 growth, which lead to the realization that 
# 2017 and 2018 data was identical and brings into question its reliability

df4 = growthdf.groupby(['Year']).agg(Average_Growth = ('Growth', 'mean')).reset_index()
df4
```

```python
# Illustrates average growth rates for below $14000 (false) and above it (true)

df5 = growthdf.groupby([growthdf["Spending"] > 14000]).agg(Average_Growth = ('Growth', 'mean')).reset_index()
df5
```
